```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import requests
```

```python
from math import sqrt
def dist(pt1, pt2):
    """Calculate the Euclidean distance between two points."""
    return sqrt((pt1[0]-pt2[0])**2 + (pt1[1]-pt2[1])**2)
```

There will be three sets of data necessary:  the station codes (`stns`), station locations (`locs`), and station temperatures (`temps`).

Load the station locations for weighted averaging.

```python
# Get latitude and longitude of stations.
# http://www.nws.noaa.gov/mdl/gfslamp/docs/stations_info.shtml

#   Load raw data as HTML string.
r = requests.get('http://www.nws.noaa.gov/mdl/gfslamp/docs/stations_info.shtml')

#   Find Illinois data segment (in a PRE tag).  We know (from examination) that inside of
#   the PRE block containing ' IL ' (with whitespace and case matching) we can find the IL
#   station data.  This solution isn't robust, but it's good enough for practical cases.
il_start  = r.text.find(' IL ')
tag_start = r.text.rfind('PRE', il_start-200, il_start) # The string.rfind() function looks backwards from an index.
tag_end   = r.text.find('PRE', il_start)
data = r.text[tag_start+4:tag_end-2]
#print(data)

#   Extract latitude and longitude of stations.  We know the columns are fixed (which is
#   both inconvenient and convenient).  In this case, we will simply set the limits of
#   the relevant columns by counting the number of columns over we need to go.
r_stn  = (5, 9) #remember that the last index is an exclusive bound
r_name = (10, 31)
r_lat  = (36, 41) #we don't need the N/W designation; we know where we are
r_lon  = (46, 51)
stns = []
locs = []
locs_x = []; locs_y = []
for line in data.split('\n'):
    stn  = line[r_stn[0]:r_stn[1]]
    name = line[r_name[0]:r_name[1]]
    try:
        lat  =  float(line[r_lat[0]:r_lat[1]])
        lon  = -float(line[r_lon[0]:r_lon[1]])
    except ValueError:
        # skip rows with invalid data
        continue
    stns.append(stn)
    locs.append((lon, lat))
    locs_x.append(lon)
    locs_y.append(lat)
```

Load the corresponding temperature data from an online source.

```python
# Retrieve temperature data from online source.
# http://www.nws.noaa.gov/mdl/gfslamp/lavlamp.shtml

# Load raw data as HTML string.
r = requests.get('http://www.nws.noaa.gov/mdl/gfslamp/lavlamp.shtml')

#   We have a list of Illinois stations from the sites loaded previously.  We need to load
#   the data for each of those sites and store these data locally.  There are a lot of data
#   included here, but we are only interested in one:  the current temperature, located at
#   the index offset 169 and of length 2 (found by examination).
temps = []
for stn in stns:
    tag_start = r.text.find(stn)
    if tag_start == -1:
        temps.append(float('NaN'))
        continue
    tag_end = tag_start + 1720 #each text block is 1720 characters long
    temps.append(float(r.text[tag_start+169:tag_start+172]))
```

Plot the temperature data thus loaded.

```python
import matplotlib as mpl
mpl.rcParams['figure.figsize']=[10,10]

#These are the stations we loaded above.
plt.scatter(locs_x, locs_y, c=temps, marker='D') #scatter plots the points and color-codes them by the c array

#Output the name of each station
for i in range(len(stns)):
    tmp = locs[i]
    plt.text(tmp[0]+0.04, tmp[1]+0.02, stns[i])

#The borders of IL in lat/lon coordinates are located in this file and included for context.
x = []; y = []
for line in open('borders_il.txt', 'r'):
    fields = line.strip().split(' ')
    x.append(float(fields[0]))
    y.append(float(fields[1]))
plt.plot(x, y, 'k-')

plt.axis('equal')
plt.title('Weather Stations in Illinois')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

Set up the interpolation.

```python
# Some sites to check:
#   Bishop Hill 41°12′N 90°7′W
#   
pt = (-90.117, 41.2)

# Find the three nearest neighbors of the point we are at.
#   Get a list of all neighbors from this point.  Store them by distance and index.
dists = []
for i in range(0, len(locs)):
    dists.append((dist(locs[i],pt), i))

#   Sort the list to take the nearest three sites.
dists.sort()
```

Interpolate some points.  Plot a complete interpolation.

```python
# Use a simple 3-nearest-neighbor weighted average rather than triangulation.
wts = [0]*3
wts[0] = 1/(dists[0][0] ** 2)
wts[1] = 1/(dists[1][0] ** 2)
wts[2] = 1/(dists[2][0] ** 2)

T_pt = (wts[0]*temps[dists[0][1]] + wts[1]*temps[dists[1][1]] + wts[2]*temps[dists[2][1]]) / sum(wts)
print(T_pt)
```
